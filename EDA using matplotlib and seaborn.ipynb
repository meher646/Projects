**Relevant Attributes**: VEHICLE YEAR, VEHICLE ENGINE HORSEPOWER, VEHICLE ANNUAL MILES TRAVELED IN LAST YEAR, VEHICLE FUEL TYPE

- **Environmental Impact Assessment**: Calculating the fleet's environmental impact according to the vehicle fuel type, vehicle age, and consumption trends of the vehicles.
**Relevant Attributes**: VEHICLE FUEL TYPE, VEHICLE ENGINE HORSEPOWER, VEHICLE YEAR, VEHICLE GROSS VEHICLE WEIGHT

# Numerical Attributes: histogram

### matplotlib:

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(Data['VEHICLE YEAR'].dropna(), bins=10, edgecolor='white')
plt.title('Distribution of Vehicle Year')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.show()
```

### seaborn:

```python
import seaborn as sns
```

```python
sns.histplot(Data['VEHICLE YEAR'].dropna(), bins=10, kde=True)
plt.title('Distribution of Vehicle Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

The frequency of vehicles is clearly trending upward. There are more cars as we go from the left (1980s) to the right (2020s).This is a result of an increase in new car sales.The rising tendency may be caused by a number of factors, including rising new car sales and production over time.

```python
avg_age_by_engine_make = Data.groupby('VEHICLE ENGINE MAKE')['Vehicle Age'].mean()


plt.figure(figsize=(10, 7))
plt.bar(avg_age_by_engine_make.index, avg_age_by_engine_make.values)
plt.xticks(rotation=75)
plt.xlabel('VEHICLE ENGINE MAKE')
plt.ylabel('Vehicle Age')
plt.show()
```

```python
plt.figure(figsize=(10, 7))
sns.barplot(x=avg_age_by_engine_make.index, y=avg_age_by_engine_make.values)
plt.xticks(rotation=75)
plt.xlabel('VEHICLE ENGINE MAKE')
plt.ylabel('Vehicle Age')
plt.show()
```

```python
avg_age_by_make = Data.groupby('VEHICLE MAKE')['Vehicle Age'].mean().sort_values()

# Plotting with Matplotlib
plt.figure(figsize=(10, 7))
plt.bar(avg_age_by_make.index, avg_age_by_make.values)
plt.xticks(rotation=75)
plt.xlabel('VEHICLE MAKE')
plt.ylabel('Average Vehicle Age')
plt.show()
```

```python
avg_age_by_make_df = avg_age_by_make.reset_index()

plt.figure(figsize=(10, 7))
sns.barplot(x='VEHICLE MAKE', y='Vehicle Age', data=avg_age_by_make_df, order=avg_age_by_make_df['VEHICLE MAKE'])
plt.xticks(rotation=75)
plt.xlabel('VEHICLE MAKE')
plt.ylabel('Average Vehicle Age')
plt.show()
```

The average age of cars by make is displayed in the bar plot. With the youngest average vehicle age, Subaru may be a brand new addition to the market or its owners may be replacing their vehicles more regularly. Conversely, White/GMC has the oldest average age, which may be a sign of longevity or a lower rate of owner change. The differing average ages between makes could also be a result of market penetration, the lifespan of a particular manufacturer, or the popularity of a brand over time. The gradient color scheme highlights the shift in average vehicle age from younger to older across various vehicle brands.

# Conclusion 

Our understanding of fleet compositions and operating efficiencies is informed by the following significant patterns that were found in the Licensees and Registrants Fleet Information dataset analysis, which focused on the relationship between vehicle fuel type, engine horsepower, and vehicle age:

- **Vehicle Year Distribution**: The dataset most likely represents a modern vehicle fleet with continuous growth or turnover, as evidenced by the increasing trend in the frequency of newer vehicles, the majority of which are from the most recent years.

- **Fuel Type Distribution**: The dataset's emphasis on business vehicles or local preferences may be the reason why diesel is by far the most popular fuel type. While there is some representation of gas vehicles, it is not as strong as diesel.

- **Type of Ownership**: Since most cars are owned rather than financed or leased, this may indicate stable finances or a preference for ownership over other ways to get a car.

- **Vehicle Body Type**: The most common vehicle body type is a roll-off truck, but there is a considerable range in body forms. 

- **Side Guard Presence**: The ratio of cars with and without side guards is nearly equal, which may indicate a change in safety regulations or a mix of older and newer models.

- **Average Vehicle Age by Make**: The average age of cars varies greatly between makes. Some brands have younger fleets, maybe as a result of recent market entry or high fleet turnover, while others have larger average ages, which might be a sign of long-term brand loyalty or longer vehicle lifespans.


The analysis concludes by highlighting the important influence that fuel type has on fleet characteristics and operational efficiency. Diesel cars have advantages in terms of power and endurance, but as the market and regulatory environments change, fleet compositions and strategies may need to be reevaluated in light of the shift towards more environmentally friendly and sustainable options.
