### Data Preprocessing

```python
data = df.copy()
```

```python
from sklearn.preprocessing import LabelEncoder
lencoders = {}
for col in data.select_dtypes(include=['object']).columns:
    lencoders[col] = LabelEncoder()
    data[col] = lencoders[col].fit_transform(data[col])
```

```python
data
```

```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
train = data[~((data < (Q1 - 1.5 * IQR)) |(data > (Q3 + 1.5 * IQR))).any(axis=1)]
train.shape
```

```python
from sklearn import preprocessing

r_scaler = preprocessing.MinMaxScaler()
r_scaler.fit(train)

train = pd.DataFrame(r_scaler.transform(train), columns=train.columns)
train.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
y = train['satisfaction']
X = train.drop(['satisfaction'],axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
# sd = StandardScaler()

# X_train = sd.fit_transform(X_train)
# X_test = sd.transform(X_test)
```

```python
from sklearn.metrics import accuracy_score,log_loss,precision_score,roc_auc_score,confusion_matrix,classification_report
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from catboost import CatBoostClassifier
from xgboost import XGBClassifier
```

```python
def classfication_results(model,X_train,X_test,y_train,y_test):
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    print(confusion_matrix(y_test,y_pred))
    print(classification_report(y_test,y_pred,digits=5))
```

### Logistic Regression

```python
model_lr = LogisticRegression()
classfication_results(model_lr,X_train,X_test,y_train,y_test)
```

### Decision Tree

```python
model_dt = DecisionTreeClassifier()
classfication_results(model_dt,X_train,X_test,y_train,y_test)
```

### Random Forest Classifier

```python
model_rf = RandomForestClassifier()
classfication_results(model_rf,X_train,X_test,y_train,y_test)
```

### Cat Boost Classifier

```python
model_cat = CatBoostClassifier(silent=True)
classfication_results(model_cat,X_train,X_test,y_train,y_test)
```

### XGB Classifier

```python
model_xgb = XGBClassifier()
classfication_results(model_xgb,X_train,X_test,y_train,y_test)
```

### Feature Importance

```python
feature_importances = model_cat.feature_importances_

# Get feature names
feature_names = model_cat.feature_names_

# Sort feature importances in descending order
sorted_idx = feature_importances.argsort()[::-1]

# Plot feature importances
plt.figure(figsize=(10, 6))
plt.bar(range(len(feature_importances)), feature_importances[sorted_idx])
plt.xticks(range(len(feature_importances)), [feature_names[i] for i in sorted_idx], rotation=90, ha='right')
plt.xlabel('Feature')
plt.ylabel('Importance')
plt.title('Feature Importances')
plt.show()
```
